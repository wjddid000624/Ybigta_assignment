군집화는 특정 목적을 위해 군집의 유형을 나누고 정의한 뒤 특징을 요약하여 활용하는 방식으로, 예시 모델에는 K-means clustering이 있습니다. K-means clustering은 K개의 cluster로 모델을 군집화하는 알고리즘입니다. 군집의 중심점이 될 데이터를 K개를 랜덤으로 고릅니다. 이후, 다른 데이터들을 그 중심점들 중에 가장 가까운 군집에 포함시키고, 그 군집의 평균과 가장 가까운 데이터를 중심점으로 삼는것을 반복한다. 중심점의 변경이 없을 때까지 중심점의 변경을 반복한다.

차원축소는 원래 데이터셋이 고차원에 표현되어 있었을 때 의미 있는 속성을 이용해서 저차원 공간에 표현하는 방식으로, 예시 모델에는 PCA가 있습니다.

```python
# 데이터셋 불러오기
import numpy as np
import pandas as pd
df = pd.read_csv('diabetes.csv')
```

```python
df.isna().sum()
```

```python
# 모델 라이브러리 import
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
# base dataframe
df_base=df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']]
```

```python
df_base.dropna(inplace=True)
```

```python
df_base
```

```python
# Model build
dt_clf = DecisionTreeClassifier(random_state=111)

# Train/test split 진행 (7:3)
X_train, X_test, y_train, y_test = train_test_split(df_base.drop('Outcome',axis=1),
                                                    df_base['Outcome'], # target variable = Outcome
                                                    test_size=0.3, random_state=111)
# 모델 훈련
dt_clf = dt_clf.fit(X_train,y_train)
```

```python
X_train
```

```python
# 평가지표 출력
from sklearn.metrics import accuracy_score
pred = dt_clf.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print(accuracy)
```
