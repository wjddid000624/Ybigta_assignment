## 과제 1

**튜플**을 오름차순 정렬해서 리스트로 반환하는 함수를 만들어주세요.
```sorted()``` 함수는 금지입니다. 버블 정렬, 삽입 정렬, 퀵 정렬 등 원하는 정렬 알고리즘을 선택해주세요


```python3
tup = (3, 4, 1)
lst = sort(tup)
# [1, 3, 4]가 나와야 함
print(lst)
```

````python
def sort(tup):
    # 이 부분은 지우지 말아주세요
    if not isinstance(tup, tuple):
        raise TypeError("tuple을 넣어주세요")
    # pass를 지우고 코드를 채워주세요
    arr = list(tup)
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                tmp = arr[j]
                arr[j] = arr[j+1]
                arr[j+1] = tmp
    
    return tuple(arr)
````

````python
# 결과확인
tup = (8, 9, 1, 3, 11)
print(sort(tup))

# 코드 한 줄 설명: (짧은 설명을 적어주세요!) bubble sort를 이용했습니다. tuple은 변경이 불가능하기 때문에 tuple을 list형태로 바꿔서 sort작업을 진행했습니다.
````

## 과제 2

1. 'Point' 라는 이름의 클래스를 만들어 주세요.

2. 'Point' 클래스는 0이 아닌 정수 x와 y를 인스턴스 변수를 생성자에 입력 받습니다.

3. 'Point' 클래스는 위치를 set하는 함수 ```set_position(x, y)```이 있습니다.

4. 'Point' 클래스는 위치를 get하는 함수 ```get_position()```이 있습니다.

5. 'Point' 클래스는 다른 Point instance와의 거리를 계산하는 ```get_dist()``` 함수가 있습니다.

6. 'Point' 클래스는 ```print(Point)```를 했을 때 Point(x: x, y: y)를 출력해야합니다.

````python
from math import sqrt

class Point:
    # pass를 지우고 클래스를 완성해주세요!!
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def get_point(self):
        print("x: " + str(self.x) + ", y: " + str(self.y))
    
    def set_point(self, x, y):
        self.x = x
        self.y = y

    def get_dist(self, point):
        return( sqrt( (self.x - point.x)**2 + (self.y - point.y)**2 ) )
````

````python
# 출력
# x: 1, y: 2
p1 = Point(1, 2)
p1.get_point()

# 출력
# x: 10, y: 20
p1.set_point(10,20)
p1.get_point()

# (10, 20)과 (3, 4)의 유클리디안 거리를 계산해주세요
p2 = Point(3, 4)
print(p1.get_dist(p2))

print(p1)
print(p2)

# 코드 한 줄 설명: (짧은 설명을 적어주세요!)
# init 함수를 통해 x와 y를 설정
# get_point 함수는 자신의 x와 y를 format에 맞춰 출력
# set_point 함수는 자신의 x와 y를 변경
# get_dist 함수에서는 두 점 사이의 거리를 구해 return
````
